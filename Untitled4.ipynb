## Data Preparation

```python
from torchvision import transforms
#استوردنا الترانفورمز عشان نسوي تعديلات على الصور

# ImageNet stats (مهمة لأن ResNet18 متدرب على ImageNet)
mean = [0.485, 0.456, 0.406] #ما نغيرها لأن ResNet18 متدرب على صور ImageNet بهذه الإحصائيات
std  = [0.229, 0.224, 0.225]

data_transforms = {
    "train": transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(), #هنا يحولها لتنسور
        transforms.Normalize(mean, std),
    ]),
    "val": transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean, std),
    ]),
}
```

## Dataset and DataLoader

```python
import torch
from torch.utils.data import IterableDataset
from PIL import Image
import tensorflow_datasets as tfds

# هذا الكلاس وظيفته:
# يحول الداتاسيت اللي جاية من TensorFlow (tfds)
# إلى Dataset يفهمه PyTorch
# بدون ما نحمل كل البيانات في الذاكرة (عشان ما يطيح الرام)
class TFDSWrapper(IterableDataset):

    def __init__(self, tfds_dataset, transform=None, shuffle=False):
        # نخزن الداتاسيت الأصلية من tfds
        self.tfds_dataset = tfds_dataset

        # نخزن الترانسفورمز (train أو val)
        self.transform = transform

        # هل نبي نخلط البيانات أو لا
        self.shuffle = shuffle

    def __iter__(self):
        # لو كنا في التدريب نخلط البيانات
        if self.shuffle:
            ds = self.tfds_dataset.shuffle(
                2048,  # حجم البفر (رقم مناسب)
                reshuffle_each_iteration=True
            )
        else:
            ds = self.tfds_dataset

        # نقرأ البيانات صورة صورة
        for img_np, label in tfds.as_numpy(ds):

            # نحول الصورة من numpy array إلى PIL Image
            # لأن torchvision.transforms تشتغل على PIL
            img = Image.fromarray(img_np)

            # نطبق الترانسفورمز لو كانت موجودة
            if self.transform:
                img = self.transform(img)

            # نرجع الصورة (Tensor) والليبل (رقم الكلاس)
            yield img, int(label)
```

```python
from torch.utils.data import DataLoader

# نبني Dataset التدريب
# نستخدم ترانسفورمز التدريب (فيها augmentation)
# ونفعل shuffle داخل الكلاس عشان البيانات تكون مخلوطة
train_dataset = TFDSWrapper(
    ds_train,
    transform=data_transforms["train"],
    shuffle=True
)

# نبني Dataset الفالديشن
# بدون augmentation عشان التقييم يكون ثابت
# وبدون shuffle عشان التقييم يكون عادل وثابت
val_dataset = TFDSWrapper(
    ds_val,
    transform=data_transforms["val"],
    shuffle=False
)

batch_size = 16
# عدد الصور اللي تدخل المودل في كل خطوة
# خليناه 16 بدل 32 عشان نقلل الضغط على الرام

num_workers = 0
# مهم جدًا مع tfds streaming
# إذا خليتيه 2 ممكن يصير crash أو مشاكل multiprocessing

# DataLoader للتدريب
train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    num_workers=num_workers
    # ملاحظة: ما نحط shuffle هنا لأن IterableDataset ما يعتمد عليه
)

# DataLoader للفالديشن
val_loader = DataLoader(
    val_dataset,
    batch_size=batch_size,
    num_workers=num_workers
)

# Quick Check: نتأكد شكل الباتش
images, labels = next(iter(train_loader))
print("Batch images shape:", images.shape)  # لازم [16, 3, 224, 224]
print("Batch labels shape:", labels.shape)  # لازم [16]
print("First labels:", labels[:10])
```

## Model Building


```python
import torch
import torch.nn as nn
from torchvision import models

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("device:", device)

num_classes = 38  # عدد الكلاسات

weights = models.ResNet18_Weights.DEFAULT
model = models.resnet18(weights=weights)

# نغير آخر طبقة عشان تناسب 38 كلاس
model.fc = nn.Linear(model.fc.in_features, num_classes)

model = model.to(device)
print("Final layer:", model.fc)
```

##Training

```python
def train_one_epoch(model, loader, criterion, optimizer, device):
    # نحط المودل في وضع التدريب
    # هذا مهم عشان الطبقات مثل Dropout و BatchNorm تشتغل صح
    model.train()

    running_loss = 0.0          # نجمع فيه مجموع الخسارة لكل الباتشات
    running_correct = 0         # نحسب كم توقع صح
    total = 0                   # عدد الصور اللي دخلت المودل

    # نلف على كل batch في DataLoader
    for inputs, labels in loader:
        # ننقل الصور والليبلز للـ GPU (أو CPU إذا ما فيه GPU)
        inputs, labels = inputs.to(device), labels.to(device)

        # نصفر النتايح القديمة
        # لازم نسويها كل مرة عشان ما تتراكم
        optimizer.zero_grad()

        # forward pass
        # ندخل الصور للمودل ويطلع التوقعات
        outputs = model(inputs)

        # نحسب الخطأ (Loss)
        loss = criterion(outputs, labels)

        # backward pass
        # نرجع الخطأ للخلف ونحسب التدرجات
        loss.backward()

        # نحدث الأوزان باستخدام الـ optimizer
        optimizer.step()

        # نطلع الكلاس المتوقع (أكبر قيمة)
        preds = outputs.argmax(dim=1)

        # نجمع الخسارة مضروبة في عدد الصور
        running_loss += loss.item() * inputs.size(0)

        # نحسب كم توقع كان صحيح
        running_correct += (preds == labels).sum().item()

        # نزيد العداد بعدد الصور في هذا الباتش
        total += labels.size(0)

    # نرجع متوسط الخسارة والدقة لهذا الإيبوك
    return running_loss / total, running_correct / total
```

```python
@torch.no_grad()
def evaluate(model, loader, criterion, device):
    # نحط المودل في وضع التقييم
    # هذا يعطّل Dropout ويثبت BatchNorm
    model.eval()

    running_loss = 0.0
    running_correct = 0
    total = 0

    # نلف على بيانات الفالديشن
    for inputs, labels in loader:
        # ننقل البيانات للـ GPU
        inputs, labels = inputs.to(device), labels.to(device)

        # forward فقط (بدون backward)
        outputs = model(inputs)

        # نحسب الخسارة
        loss = criterion(outputs, labels)

        # نطلع التوقعات
        preds = outputs.argmax(dim=1)

        # نجمع الخسارة
        running_loss += loss.item() * inputs.size(0)

        # نجمع عدد التوقعات الصحيحة
        running_correct += (preds == labels).sum().item()

        # نزيد عدد العينات
        total += labels.size(0)

    # نرجع متوسط الخسارة والدقة
    return running_loss / total, running_correct / total
```

```python
def fit(model, train_loader, val_loader, criterion, optimizer, device, epochs=2):
    # نخزن تاريخ التدريب عشان نقدر نرسمه بعدين
    history = {
        "train_loss": [],
        "train_acc": [],
        "val_loss": [],
        "val_acc": []
    }

    # نلف على عدد الإيبوكس
    for epoch in range(epochs):

        # تدريب إيبوك واحد
        train_loss, train_acc = train_one_epoch(
            model, train_loader, criterion, optimizer, device
        )

        # تقييم على بيانات الفالديشن
        val_loss, val_acc = evaluate(
            model, val_loader, criterion, device
        )

        # نخزن النتائج
        history["train_loss"].append(train_loss)
        history["train_acc"].append(train_acc)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        # نطبع ملخص الإيبوك
        print(
            f"Epoch {epoch+1}/{epochs} | "
            f"train_loss={train_loss:.4f} train_acc={train_acc:.4f} | "
            f"val_loss={val_loss:.4f} val_acc={val_acc:.4f}"
        )

    # نرجع كل التاريخ
    return history
```

```python
def fit(model, train_loader, val_loader, criterion, optimizer, device, epochs=2):
    # نخزن تاريخ التدريب عشان نقدر نرسمه بعدين
    history = {
        "train_loss": [],
        "train_acc": [],
        "val_loss": [],
        "val_acc": []
    }

    # نلف على عدد الإيبوكس
    for epoch in range(epochs):

        # تدريب إيبوك واحد
        train_loss, train_acc = train_one_epoch(
            model, train_loader, criterion, optimizer, device
        )

        # تقييم على بيانات الفالديشن
        val_loss, val_acc = evaluate(
            model, val_loader, criterion, device
        )

        # نخزن النتائج
        history["train_loss"].append(train_loss)
        history["train_acc"].append(train_acc)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        # نطبع ملخص الإيبوك
        print(
            f"Epoch {epoch+1}/{epochs} | "
            f"train_loss={train_loss:.4f} train_acc={train_acc:.4f} | "
            f"val_loss={val_loss:.4f} val_acc={val_acc:.4f}"
        )

    # نرجع كل التاريخ
    return history
```

```python
import torch
import torch.nn as nn
from torchvision import models
# استوردنا PyTorch والموديلات الجاهزة من torchvision

# نحدد الجهاز
# إذا فيه GPU نستخدمه، غير كذا نشتغل على CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

num_classes = 38
# عدد الكلاسات في داتاسيت أمراض النباتات (Plant Disease)

# نستخدم الأوزان الافتراضية لـ ResNet18
# المودل هذا متدرب مسبقًا على ImageNet
weights = models.ResNet18_Weights.DEFAULT

# نبني مودل ResNet18 مع الأوزان الجاهزة
model = models.resnet18(weights=weights)

# نغير آخر طبقة (Fully Connected Layer)
# لأن عدد الكلاسات عندنا مختلف عن ImageNet
# ImageNet = 1000 class
# مشروعنا = 38 class
model.fc = nn.Linear(model.fc.in_features, num_classes)

# ننقل المودل للجهاز (GPU أو CPU)
model = model.to(device)

# نطبع الجهاز للتأكد إن المودل فعلاً على الـ GPU
print("Model ready on:", next(model.parameters()).device)
```

```python
# نجمد كل الطبقات
for p in model.parameters():
    p.requires_grad = False

# نفك تجميد fc فقط
for p in model.fc.parameters():
    p.requires_grad = True

criterion = nn.CrossEntropyLoss() #حساب اللوس
optimizer = torch.optim.Adam(model.fc.parameters(), lr=1e-3) #الاوبتمايزر وحددت الليرننق ريت

history_feat = fit(model, train_loader, val_loader, criterion, optimizer, device, epochs=2)
```

```python
model_ft = models.resnet18(weights=weights)
model_ft.fc = nn.Linear(model_ft.fc.in_features, num_classes)
model_ft = model_ft.to(device)

for p in model_ft.parameters():
    p.requires_grad = True

criterion_ft = nn.CrossEntropyLoss()
optimizer_ft = torch.optim.Adam(model_ft.parameters(), lr=1e-4)

history_ft = fit(model_ft, train_loader, val_loader, criterion_ft, optimizer_ft, device, epochs=2)
```

```python
print("Feature Extraction:")
#هنا جمل طباعة عشان اقارن بين الtwo models
#واشوف مين كان افضل في التريينق
print("  Best Val Acc :", max(history_feat["val_acc"]))
print("  Final Val Acc:", history_feat["val_acc"][-1])

print("\nFine-tuning:")
print("  Best Val Acc :", max(history_ft["val_acc"]))
print("  Final Val Acc:", history_ft["val_acc"][-1])
```

```python
import matplotlib.pyplot as plt

def plot_loss(history, title="Loss"):
    plt.figure(figsize=(7,4))
    plt.plot(history["train_loss"], label="train_loss")
    plt.plot(history["val_loss"], label="val_loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

plot_loss(history_feat, title="Feature Extraction - Loss")
plot_loss(history_ft,   title="Fine-tuning - Loss")
```
